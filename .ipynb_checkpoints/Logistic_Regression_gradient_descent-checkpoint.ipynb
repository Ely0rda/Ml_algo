# Gradient descent for logistic regression

```python
import copy, math
import numpy as np
import matplotlib.pyplot as plt
from kobeni import *
```

### Data set

```python
x_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train = np.array([0, 0, 0, 1, 1, 1])
```

### Plotting data

```python
pos = y_train == 1
neg = y_train == 0
pos = pos.reshape(-1,)
neg = neg.reshape(-1,)
# print(f"neg : {neg}")
fig,ax = plt.subplots(1,1,figsize=(4,4))
ax.set_ylabel(r'$x_1$')
ax.set_xlabel(r'$x_0$')
ax.scatter(x_train[pos,0],x_train[pos, 1], marker='x',c ='red')
ax.scatter(x_train[neg,0],x_train[neg, 1], marker='o',c ='blue')
ax.axis([0, 4, 0, 3.5])
plt.legend(['y=1','y=0'])
plt.show()
```

### Sgimoid function

```python
def sigmoid(x,w,b) :
    z = compute_fwb(x,w,b)
    s = 1/(1+np.exp(-z))
    return s
```

### Runing Gradient descent

```python
w_in = np.zeros(x_train.shape[1])
b_in = 0.0
alpha = 0.1
num_iter = 10000
w_final,b_final,_ = compute_gd(x_train,y_train,w_in,b_in,num_iter,alpha,1000,sigmoid)
```

### Plotting the result of gradient descent

```python
pos = y_train == 1
neg = y_train == 0
pos = pos.reshape(-1,)
neg = neg.reshape(-1,)
# print(f"neg : {neg}")
fig,ax = plt.subplots(1,1,figsize=(5,4))
ax.set_ylabel(r'$x_1$')
ax.set_xlabel(r'$x_0$')
ax.scatter(x_train[pos,0],x_train[pos, 1], marker='x',c ='red')
ax.scatter(x_train[neg,0],x_train[neg, 1], marker='o',c ='blue')
ax.axis([0, 4, 0, 3.5])
plt.legend(['y=1','y=0'])
#Drawing the decisoin boundary
#We need tw0 points
#The decisoin boundary is when w1*x1+w2*x2 + b = 0
#The function for decisoin boundaey will look like this
#w1*x1 = -b-w2*x2
#x1 = (-b-w2*x2)/w1
#x1 = -b/w1 - (w2/w1)*x2
#The first point we can get is x1 when x2=0
#x1 = -b/w1
#Point1 : [0,-b/w1]
#The second point we can get is x2 when x1=0
#x2 = -b/w2 
#Point2 : [-b/w2,x2]
x0 = -b_final/w_final[0]
x1 = -b_final/w_final[1]
ax.axis([0,4,0,3.5])
ax.plot([0,x0],[x1,0],c='orange',lw=1)
plt.show()
```
