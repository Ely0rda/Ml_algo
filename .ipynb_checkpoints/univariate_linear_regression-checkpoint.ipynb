```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

### Training data

```python
x_train = np.array([1.0,2.0])
y_train = np.array([300.0, 500.0])
```

### Visualizing the model
- We need a function that takes w,b,x and return y-hat

```python
def compute_fwb(x,w,b) :
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = x[i]*w+b
    return f_wb
```

## Training the model using gradient descent
    - A function for calculating the derivatives
    - implement gradient descent
        

### Calculating the cost function
    The cost function here will be a good measure for the progress of
    gradient descent

```python
def cost_function(w,b,x,y):
    fwb = compute_fwb(x,w,b)
    for i in range(x):
        p = pow(fwb[i]-y[i],2)
        cost += p*0.5*x.shape[0]
    return cost
```

### Calculating the derivatives
    Every time we got a new w,b from the gradient descent we need to calculate the derivatives

```python
def derivatives(w,b,x,y):
    m = x.shape[0]
    f_wb = compute_fwb(x,w,b)
    dj_dw =0.0
    dj_db =0.0
    for i in range(m):
        dj_dw += (f_wb[i]-y[i])*x[i]
        dj_db += (f_wb[i]-y[i])
    dj_dw /= m
    dj_db /= m
    return dj_dw,dj_db
    
```

### Applying gradient descent

```python
def compute_gd(w_in, b_in,x,y,alpha,num_iter) :
    m = x.shape[0]
    w = w_in
    b= b_in
    J_history = []
    for i in range(num_iter):
        dj_dw, dj_db = derivatives(w,b,x,y)
        w = w - alpha*dj_dw
        b = b - alpha*dj_db
        J_history.append(cost_function(w,b,x,y))
        if i%1000 == 0 :
            print(f"w ={w} ,b ={b}, dw ={dj_dw}, db={dj_db}")
    return w,b,J_history
   
```

```python
w,b,J_history = compute_gd(0.0,0.0,x_train,y_train,1.0e-2,10000)
```

## Plotting

### Cost function versus number of iterations

```python
plt.plot(np.arrange(len(J_history)),J_history)
plt.xlabel('Number of iterations')
plt.ylabel('Cost function')
plt.title('cost vs number of iterations')
```
